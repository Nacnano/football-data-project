UEFA_Ranking 
- the score of each football club calculate by UEFA plus the country part score obtained by overall national team performance
- we can use this to map the player national to this meaningful score

```python
import pandas as pd

for year in range(2018,2027):
    print(year)
    url = f"https://kassiesa.net/uefa/data/method5/trank10-{year}.html"
    tables = pd.read_html(url)


    df = tables[0]         
    df.drop(columns=['Unnamed: 0','Unnamed: 1'],inplace=True)
    df.rename(columns={"Unnamed: 2": "team", "Unnamed: 3": "nationality"},inplace=True)
    df.dropna(subset=['Total Points'],inplace=True)
    df.reset_index(drop=True,inplace=True)
    df.to_csv(f'../data/uefa/UEFA_Ranking_{year}.csv',index=False)
    print(df.head(2))
```
