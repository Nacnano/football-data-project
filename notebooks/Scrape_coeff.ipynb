```python
import pandas as pd

url = "https://kassiesa.net/uefa/data/method5/trank10-2026.html"
tables = pd.read_html(url)

print(len(tables))      # shows how many tables found
df = tables[0]          # pick the first one
df.drop(columns=['Unnamed: 0','Unnamed: 1'],inplace=True)
df.rename(columns={"Unnamed: 2": "team", "Unnamed: 3": "nationality"},inplace=True)
df.dropna(subset=['Total Points'],inplace=True)
df.reset_index(drop=True,inplace=True)
df.to_csv('../data/UEFA_Ranking_2026.csv',index=False)
print(df.head())
```
